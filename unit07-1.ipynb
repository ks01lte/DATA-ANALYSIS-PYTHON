```python
# 0세부터 100세 이상까지의 데이터, 즉 총 101개의 데이터인지 확인
import csv
f = open('age.csv')
data = csv.reader(f)

for row in data :
    if '신도림' in row[0] :
        print(len(row[3:]))
```

```python
# 0세부터 100세 이상의 인구수 리스트에 저장
import csv
f = open('age.csv')
data = csv.reader(f)
result = []

for row in data :
    if '신도림' in row[0] :
        for i in row[3:]:
            result.append(int(i))

import matplotlib.pyplot as plt
plt.style.use('ggplot') #격자무늬스타일지정
plt.plot(result)
plt.show()
```

```python
# 0세부터 100세 이상의 인구수 리스트에 저장
import csv
f = open('age.csv')
data = csv.reader(f)
result = []
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요 : ')

for row in data :
    if name in row[0] :
        name = row[0]
        for i in row[3:]:
            result.append(int(i.replace(',','')))

import matplotlib.pyplot as plt
plt.style.use('ggplot') #격자무늬스타일지정
plt.plot(result)
plt.rc('font', family='Malgun Gothic')
plt.title(name + ' 지역의 인구 구조')
plt.show()
```
