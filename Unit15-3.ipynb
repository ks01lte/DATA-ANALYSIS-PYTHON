```python
# +궁금한 지역 이름 입력받고 해당 지역의 인구구조 저장하기

import pandas as pd
df = pd.read_csv('age.csv', encoding='cp949', index_col=0)
df = df.div(df['2019년02월_계_총인구수'], axis=0) # 전체 데이터를 총인구수로 나눠서 비율로 환산
del df['2019년02월_계_총인구수'], df['2019년02월_계_연령구간인구수'] # 총인구수, 연령구간인구수 열 삭제

name = input('원하는 지역의 이름을 입력해주세요 : ') # 지역 이름 입력
a = df.index.str.contains(name) # 해당 행을 찾아서 해당 지역의 인구 구조를 저장
df2 = df[a]
df2

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
df2.T.plot()
plt.show()
```

```python
# +궁금한 지역의 인구 구조와 가장 비슷한 인구 구조를 가진 지역 시각화하기

import pandas as pd
df = pd.read_csv('age.csv', encoding='cp949', index_col=0)
df = df.div(df['2019년02월_계_총인구수'], axis=0) # 전체 데이터를 총인구수로 나눠서 비율로 환산
del df['2019년02월_계_총인구수'], df['2019년02월_계_연령구간인구수'] # 총인구수, 연령구간인구수 열 삭제

name = input('원하는 지역의 이름을 입력해주세요 : ') # 지역 이름 입력
a = df.index.str.contains(name) # 해당 행을 찾아서 해당 지역의 인구 구조를 저장
df2 = df[a]
df2

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
df2.T.plot()
plt.show()

# A읜 인구비율에서 B인구비율을 뺀 값의 제곱 값을 모두 더한다
import numpy as np
x = df.sub(df2.iloc[0], axis=1)
y = np.power(x,2) 
z = y.sum(axis=1)

i =  z.sort_values().index[:5] #차이가 가장 작은 지역 5곳
df.loc[i].T.plot()
plt.show()
```
