# numpy array 생성하기

```python
import numpy as np
a = np.array([1,2,3,4])
print(a)
print(a[1], a[-1]) #a의 1번 인덱스 값, -1번 인덱스 값 출력
print(a[1:]) #a의 1번 인덱스를 기준을오 슬라이싱 결과 출력
```

```python
import numpy as np
a = np.array([1,2,'3',4])
print(a)
#모두 문자형으로 변환되었음을 알 수 있다. (ndarray의 경우 한 가지 데이터 타입만을 저장할 수 있다.)
```

```python
# zeros(), ones(), eye()와 같은 다양한 함수를 사용해 배열을 초기화할 수 있다.
import numpy as np

a = np.zeros(10) # 0으로 이루어진 크기가 10인 배열 생성
print(a)

a = np.ones(10) # 1으로 이루어진 크기가 10인 배열 생성
print(a)

a = np.eye(3) # 3행 x 3열의 단위 배열 생성
print(a)
```

```python
import numpy as np

# numpy의 arrange 함수에는 값을 1개에서 3개까지 입력하여 사용할 수 있다.
print(np.arange(3)) # 0 1 2
print(np.arange(3,7)) # 3 4 5 6
print(np.arange(3,7,2)) # 3 5
```

```python
import numpy as np

a=np.arange(1, 2, 0.1) # 0.1 간격으로
b=np.linspace(1,2,11) # 11개 구간으로

print(a)
print(b)
```

```python
import numpy as np
a = np.arange(-np.pi, np.pi, np.pi/10)
b=np.linspace(-np.pi, np.pi, 20)
print(a)
print(b)
```

```python
import numpy as np
a = np.zeros(10)+5
print(a)
```

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(-np.pi, np.pi, np.pi/100)
plt.plot(a, np.sin(a))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(-5, 5)
print(a)
print(a<0)
print(a[a<0])
```

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(-5, 5)
mask1 = abs(a) > 3
mask2 = abs(a) % 2 == 0

print(a[mask1])
print(a[mask1+mask2])
print(a[mask1*mask2])
```

```python
# numpy 라이브러리를 사용하여 재미 있는 버블 차트 그리기
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)
size = np.random.rand(100) * 100

mask1 = abs(x) > 50
mask2 = abs(y) > 50
x = x[mask1+mask2]
y = y[mask1+mask2]

plt.scatter(x,y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

```python
# numpy 라이브러리를 사용하여 재미 있는 버블 차트 그리기
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)


mask1 = abs(x) > 50
mask2 = abs(y) > 50
x = x[mask1+mask2]
y = y[mask1+mask2]

size = np.random.rand(len(x)) * 100

plt.scatter(x,y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
