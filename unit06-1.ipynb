```python
# 1907년부터 2018년까지 수집된 서울의 최고 기온을 히스토그램으로 표현

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[-1] != '' :
        result.append(float(row[-1]))
        
plt.hist(result, bins=100, color = 'r')
plt.show()
```

```python
# 1907년부터 2018년까지 수집된 서울의 8월 최고 기온을 히스토그램으로 표현

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == '08' :
            aug.append(float(row[-1]))
        
plt.hist(aug, bins=100, color = 'r')
plt.show()
```

```python
# 1907년부터 2018년까지 수집된 서울의 1월과 8월 최고 기온을 히스토그램으로 표현

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
jan = []
aug = []

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == '08' :
            aug.append(float(row[-1]))
        if month == '01' :
            jan.append(float(row[-1]))
        
plt.hist(aug, bins=100, color = 'r', label = 'Aug')
plt.hist(jan, bins=100, color = 'b', label = 'Jan')
plt.legend()
plt.show()
```

```python
# 기온 데이터를 상자그림으로 표현하기
# 서울의 전체 최고 기온

import csv
f = open ('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[-1] != '' :
        result.append(float(row[-1]))

import matplotlib.pyplot as plt
plt.boxplot(result) #상자그림으로 나타내기
plt.show()
```

```python
# 기온 데이터를 상자그림으로 표현하기
# 서울의 1월과 8월의 최고 기온

import csv
f = open ('seoul.csv')
data = csv.reader(f)
next(data)
jan = []
aug = []

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == '01' :
            jan.append(float(row[-1]))
        if month == '08' :
            aug.append(float(row[-1]))

import matplotlib.pyplot as plt
plt.boxplot(jan) # 1월 최고기온 상자그림으로 나타내기
plt.boxplot(aug) # 8월 최고기온 상자그림으로 나타내기
plt.show()
```

```python
# 기온 데이터를 상자그림으로 표현하기
# 서울의 1월과 8월의 최고 기온의 데이터를 원소로 하는 리스트를 상자그림으로!

import csv
f = open ('seoul.csv')
data = csv.reader(f)
next(data)
jan = []
aug = []

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == '01' :
            jan.append(float(row[-1]))
        if month == '08' :
            aug.append(float(row[-1]))

import matplotlib.pyplot as plt
plt.boxplot([jan,aug])
plt.show()
```

```python
# 기온 데이터를 상자그림으로 표현하기
# 서울의 매달마다 최고 기온의 데이터를 상자그림으로 나타낸다면?

import csv
f = open ('seoul.csv')
data = csv.reader(f)
next(data)
month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data :
    mon = row[0].split('-')[1] #int형으로 변환할 필요 있음
    if row[-1] != '' :
        month[int(mon)-1].append(float(row[-1])) #int빼먹지 않도록 주의!

import matplotlib.pyplot as plt
plt.boxplot(month)
plt.show()
```

```python
# 기온 데이터를 상자그림으로 표현하기
# 서울의 8월 일별 최고 기온의 데이터를 상자그림으로 나타낸다면?

import csv
f = open ('seoul.csv')
data = csv.reader(f)
next(data)

# 31일의 일별 최고 기온을 저장할 수 있는 리스트 생성 (반복문 활용)
day = []
for i in range(31) :
    day.append([])

for row in data :
    m = row[0].split('-')[1]
    d = row[0].split('-')[2]
    if row[-1] != '' :
        if m == '08' :
            day[int(d)-1].append(float(row[-1]))

import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.boxplot(day, showfliers=False)

plt.show()
```
