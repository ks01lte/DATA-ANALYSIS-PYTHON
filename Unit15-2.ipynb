```python
import pandas as pd
import numpy as np

index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8,3), index=index, columns=list('ABC'))
df['D'] = df['A'] / df['B']
df['E'] = np.sum(df, axis=1)
df = df.sub(df['A'], axis=0) #A열데이터 기준으로 열우선계산 (전체 데이터를 A열의 값으로 뺀다)
df = df.div(df['C'], axis=0) #C열데이터 기준으로 열우선계산 (전체 데이터를 C열의 값으로 나눈다)

df.to_csv('test.csv')
df.head()
```
